```python
import pandas as pd
import numpy as np
import math as mt
```

```python
store = pd.read_csv('Store.csv',sep=',',header=0,encoding="latin")
store
```

```python
#•	How many unique cities are the orders being delivered to
store.City.nunique()
```

```python
#•	What is the total quantity sold in the East Region?
len(store.query("(Region=='East')"))
```

```python
#•	What is the sum quantity sold in the East Region?
store.query("(Region=='East')").Quantity.sum()
```

```python
#•	In the south region sort the sales in decreasing order
sorted(store.Sales,reverse=True)
```

```python
# •	Find the mean of quantity for every region
store.groupby('Region').mean().Quantity
```

```python
#•	Find the mean of sales for every category
store.groupby('Region').mean().Sales
```

```python
# •	Find the max, min, sum of sales for every category
print("the max, min, sum of profit for every category")
mxs=store.groupby('Category').max().Sales
mns=store.groupby('Category').min().Sales
ss=store.groupby('Category').sum().Sales
print("max = ", mxs , "mns=",mns,"sum = ",ss)
print("the max, min, sum of profit for every category")
mxp=store.groupby('Category').max().Profit
mnp=store.groupby('Category').min().Profit
sp=store.groupby('Category').sum().Profit
print("max = ", mxp , "mns=",mnp,"sum = ",sp)
```

```python
#•	Find sum of sales and max profit for every segment
print("Sales = ",store.groupby('Segment').sum().Sales)
print("Profit =",store.groupby('Segment').max().Profit)
```

```python
# •	For every segment find the mean of the discount
print("Mean = ",store.groupby('Segment').mean().Discount)
```

```python
#•	For every segment find the most profitable customers
a=[store.groupby("Segment").max().Profit]
b=[store.Profit]
store[["Customer Name","Segment","Profit"]][np.in1d(b,a)]
```

```python
#•	What are the top 5 categories that give maximum profit?
store.groupby('Sub-Category').sum().Profit.head(5)
```

```python
#•	What is the Total Sales, Quantity, Discount, Profit across Total US.
store[["Sales","Quantity","Profit"]].sum()
```

```python
#•	How many times has it taken more than 5 days from placing an order to shipping

store["Ship Date"]= pd.to_datetime(store['Ship Date'])
b=store["Order Date"]= pd.to_datetime(store['Order Date'])
len(store[(store["Ship Date"]-store["Order Date"])>'5 days'])
```

```python
#•	Find the total number of orders in every category which has been shipped with a duration > 5 days
store[(store["Ship Date"]-store["Order Date"])>'5 days'].groupby("Category").count().City
```

```python
#•	What’s the percentage of items which has been shipped within 5 days
len(store[(store["Ship Date"]-store["Order Date"])<'5 days'])/len(store)*100
```

```python
#•	What’s the percentage of items which has been shipped after 5 days

len(store[(store["Ship Date"]-store["Order Date"])>'5 days'])/len(store)*100
```
