```python
# c.	For each brand having price >=2.5 find the mean, median, sd of the log of income

mn=data[data.price>=2.5].groupby(["brand"]).mean().LogInc
md=data[data.price>=2.5].groupby(["brand"]).median().LogInc
std1=data[data.price>=2.5].groupby(["brand"]).std().LogInc
print("Mean= ", mn,"Median = ",md,"Standard deviation =",std1)
```
